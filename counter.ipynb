```python
with open('1000-champs.csv', 'r') as f:
    entries = f.read().split(', ')
entries
```

```python
teams = {}
for e in entries:
    if e in teams:
        teams[e] += 1
    else:
        teams[e] = 1
teams
```

```python
for t in teams:
    teams[t] /= 10
teams
```

```python
sortedTeams = sorted(teams.items(), key=lambda kv: kv[1], reverse=True)
sortedTeams
```
